```python
import pandas as pd
import plotly.graph_objects as go

# Cargar el CSV correctamente
df = pd.read_csv("Perfilmigrante.csv", sep=';')

# Elegimos solo las columnas necesarias
cols = ['País', '0-14 (Extranjeros)', '15-64 (Extranjeros)', '65+ (Extranjeros)']
df_radar = df[cols]
df_radar = df_radar.dropna()


# Crear la figura
fig = go.Figure()

# Agregar trazas para cada país
for _, row in df_radar.iterrows():
    fig.add_trace(go.Scatterpolar(
        r=[row['0-14 (Extranjeros)'], row['15-64 (Extranjeros)'], row['65+ (Extranjeros)']],
        theta=['0-14', '15-64', '65+'],
        fill='toself',
        name=row['País'],
        marker=dict(size=8)
    ))

# Configuración estética
fig.update_layout(
    width=1000,
    height=1000,
    polar=dict(
        radialaxis=dict(visible=True, range=[0, 90], tickfont=dict(size=17)),
        angularaxis=dict(tickfont=dict(size=18)),
        domain=dict(x=[0.1, 0.9], y=[0.1, 0.9])
    ),
    title=dict(
        text="Perfil Etario de la Población Migrante - Todos los Países",
        font=dict(size=22)
    ),
    legend=dict(font=dict(size=13)),
    showlegend=True
)

fig.show()
```
