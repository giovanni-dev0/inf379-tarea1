```python
import pandas as pd
import plotly.express as px


df = pd.read_csv("DatosPobrezaLatam.csv", sep=';')


df = pd.read_csv("DatosPobrezaLatam.csv", sep=';')



# Crear columna combinada
df['Vulnerabilidad'] = df['Pobreza (%)'] + df['Desempleo (%)']

fig = px.treemap(
    df,
    path=['País'],
    values='Vulnerabilidad',
    color='IDH',
    color_continuous_scale='RdYlGn',
    hover_data={
        'Pobreza (%)': True,
        'Desempleo (%)': True,
        'IDH': True,
        'Vulnerabilidad': True
    },
    title='Vulnerabilidad Económica: (Pobreza + Desempleo) vs IDH'
)


fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))


fig.show()
```
