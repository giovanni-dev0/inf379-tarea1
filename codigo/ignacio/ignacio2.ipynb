```python
import pandas as pd
import holoviews as hv
from holoviews import opts

hv.extension('bokeh')

categorias = ['Formal', 'Informal', 'Autonomo', 'Temporal']
color_map = {
    'Formal': '#1f77b4',
    'Informal': '#d62728',
    'Autonomo': '#9467bd',
    'Temporal': '#ff7f0e'
}

data = pd.DataFrame({
    'source': ['Argentina', 'Argentina', 'Argentina', 'Argentina',
               'Costa Rica', 'Costa Rica', 'Costa Rica', 'Costa Rica',
               'Repblica Dominicana', 'Repblica Dominicana', 'Repblica Dominicana', 'Repblica Dominicana',
               'Venezuela', 'Venezuela', 'Venezuela', 'Venezuela',
               'Colombia', 'Colombia', 'Colombia', 'Colombia',
               'Chile', 'Chile', 'Chile', 'Chile',
               'Per', 'Per', 'Per', 'Per',
               'Ecuador', 'Ecuador', 'Ecuador', 'Ecuador',
               'Brasil', 'Brasil', 'Brasil', 'Brasil',
               'Mxico', 'Mxico', 'Mxico', 'Mxico'],
    'target': categorias * 10,
    'value': [
        30, 50, 10, 10,
        40, 30, 15, 15,
        10, 70, 10, 10,
        25, 55, 10, 10,
        35, 45, 10, 10,
        45, 35, 10, 10,
        30, 50, 10, 10,
        28, 52, 10, 10,
        38, 42, 10, 10,
        36, 44, 10, 10
    ]
})

chord = hv.Chord(data).opts(
    opts.Chord(
        labels='index',
        node_color='source',
        edge_color='target',
        cmap=color_map,
        edge_cmap=color_map,
        edge_color_index='target',
        width=1000,
        height=1000,
        title="Condiciones laborales de migrantes en Amrica Latina",
        label_text_font_size='8pt',
        show_frame=False
    )
)

chord
```
